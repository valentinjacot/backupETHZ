### 4.4 &ndash; Upload a file into HDFS

Let us go over some common [commands](https://hadoop.apache.org/docs/r2.4.1/hadoop-project-dist/hadoop-common/FileSystemShell.html):
```
 ls, cat, mkdir, cp, rm, rmdir, copyFromLocal, copyToLocal
```
1. List the directories or files: ```ls```
    - For a file ls returns statistics on the file: ```$hdfs dfs -ls /example/data/fruits.txt```
    <br>This gives you the following:
    ```-rw-r--r--   1 root supergroup         66 2019-03-10 15:45 /example/data/fruits.txt```
    - For a directory it returns list of its direct children as in Unix: ```$ hdfs dfs -ls /example```
    <br>This gives you the following:
    ```Found 2 items
    drwxr-xr-x   - root supergroup          0 2019-03-10 15:46 /example/data
    drwxr-xr-x   - root supergroup          0 2019-03-10 15:46 /example/jars
    ```
    - List files in a directory: ```$ hdfs dfs -ls /example/data/*```
    - List all the directories in your HDFS: ```$ hdfs dfs -ls -R /```
<br><br>
2. Copies source paths to stdout: ```cat```
    - Inspect the file content (!DO NOT do it for a large file): ```$ hdfs dfs -cat /example/data/fruits.txt```
<br><br>
3. Copy files from source to destination: ```cp```
    - ```$ hdfs dfs -cp /example/data/fruits.txt /example/data/fruits-copy.txt```
<br><br>
4. Delete files: ```rm```
    - ```$ hdfs dfs -rm /example/data/fruits-copy.txt```<br>
    This gives you:
    ```19/03/10 16:57:11 INFO fs.TrashPolicyDefault: Moved: 'wasb://susiecluster-2019-03-10t15-05-38-179z@hdfsdata.blob.core.windows.net/example/data/fruits-copy.txt' to trash at: wasb://susiecluster-2019-03-10t15-05-38-179z@hdfsdata.blob.core.windows.net/user/sshuser/.Trash/Current/example/data/fruits-copy.txt```, the file is trashed. 
    - When you then check ```$ hdfs dfs -ls /example/data/fruits-copy.txt```,
    you will see the file does not exist anymore by getting this message: 
    ```ls: '/example/data/fruits-copy.txt': No such file or directory```
<br><br>

6. Creates directories: ```mkdir```
    - We create a folder for our exercise by ```$ hdfs dfs -mkdir /ex03```
<br><br>

7. Copy a file from the file system where your terminal locates to HDFS: ```copyFromLocal```
    - First we create a file on the file system where your terminal locates (can be local or on a cluster):
    ```
    $ echo "we create a file on the cluster(hn0) file system" > cluster.txt
    $ cat cluster.txt
    ```
    - These two commands are equivalent:
    ```
    $ hdfs dfs -copyFromLocal -f local.txt /ex03/
    $ hdfs dfs -put -f cluster.txt /ex03/
    ```
    - The -f option will overwrite the destination if it already exists.
<br><br>
8. The reverse operation of ```copyFromLocal```, i.e., download a file from HDFS to the file system where your terminal locates: ```copyToLocal```

    - ```$ hdfs dfs -copyToLocal /example/data/fruits.txt```
    - If you do ```$ ls``` in your terminal, you should see the "fruits" file there.
<br><br>    

9. Delete a directory: ```rmdir```
    - We cannot delete a non-empty folder directly, so we need to remove the files inside by ```$ hdfs dfs -rm /ex03/*```
    - We then remove the folder ```$ hdfs dfs -rmdir /ex03```
    - If you do ls again on the deleted folder, it should not exist any more. ```$ hdfs dfs -ls /ex03```
